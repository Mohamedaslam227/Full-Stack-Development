```python
import os
import PIL.Image as Image
import torchvision.transforms as transforms
```

```python
path="C:/Users/21BAD051/Downloads/ChestScan"
```

```python
class ImageLoader:
    def __init__(self,path):
        self.ImagePath = path
        self.ImgTensorData=[]
    def Image_path(self):
        for dir,_,folders in os.walk(self.ImagePath):
            for file in folders:
                if file.endswith('.jpg') or file.endswith('.png') or file.endswith('.jpeg'):
                    print(os.path.join(dir, file))
    def ImageToTensorConverter(self):
        transform = transforms.Compose([transforms.Resize((224, 224)),
        transforms.ToTensor()])
        for root, _, files in os.walk(self.ImagePath):
            for file in files:
                if file.endswith('.jpg') or file.endswith('.png'):
                    image_path = os.path.join(root, file)
                    image = Image.open(image_path)
                    image = transform(image)
                    self.ImgTensorData.append(image)
    
        return self.ImgTensorData

```

```python
ImageData=ImageLoader(path)
```
